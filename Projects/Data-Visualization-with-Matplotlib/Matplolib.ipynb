```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'-r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X lable')
axes.set_ylabel('Y lable')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('X lable')
axes1.set_ylabel('Y lable')
axes1.set_title('Larger Title')

axes2.plot(y,x,'b')
axes2.set_xlabel('X lable')
axes2.set_ylabel('Y lable')
axes2.set_title('Smaller Title')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
# axes.plot(x,y) 
axes[0].plot(x,y)
axes[0].set_xlabel('X Lablel')
```

```python
fig.savefig('MY_Pic.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='WOW')
ax.plot(y,x,label='WOW2')
ax.legend(loc=0)
```
